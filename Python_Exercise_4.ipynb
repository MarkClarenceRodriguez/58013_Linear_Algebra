Vector Operations and Its Applications

```python
import numpy as np

A = np.array([-0.4, 4.3, -0.6])
B = np.array([-0.2, 0.21, 0])
C = np.array([-0.2, 2.1, -1.5])

D = A + B + C

print("Resulting vector D:", D)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x, y, z = D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x, y, z, color='r', marker='o', s=100)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
